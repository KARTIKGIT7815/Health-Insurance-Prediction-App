```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings as w
import pickle 
```

```python
w.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../datasets/Health_insurance.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns='region', inplace=True)
```

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f'Column {df[col].value_counts()}')
    print('='*50)
```

```python
df['sex_enc'] = df['sex'].map({'female':1, 'male':0})
df['smoker_enc'] = df['smoker'].map({'yes':1, 'no':0})
```

```python
df.head()
```

```python
df.columns
```

```python
X = df[['age', 'bmi', 'children', 'sex_enc', 'smoker_enc']]
y = df['charges']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
print(f'Coefficient is {model.coef_}, Intercept is {model.intercept_}')
```

```python
data = {
    "age": [34, 23, 28],
    "sex": ["male", "female", "male"],
    "bmi": [25.0, 20.0, 21.0],
    "children": [2, 0, 1],
    "smoker": ["no", "no", "yes"]
}
df_test = pd.DataFrame(data)
```

```python
df_test
```

```python
df_test['sex_enc'] = df_test['sex'].map({'female':1, 'male':0})
df_test['smoker_enc'] = df_test['smoker'].map({'yes':1, 'no':0})
```

```python
df_test
```

```python
req_col = ['age', 'bmi', 'children', 'sex_enc', 'smoker_enc']
```

```python
df_test[req_col]
```

```python
model.predict(df_test[req_col])
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
yp = model.predict(X)
ya = df['charges']
```

```python
print(f"Mean Absolute Error is {mean_absolute_error(ya,yp)}")
```

```python
print(f"Mean Squared Error is {mean_squared_error(ya,yp)}")
```

```python
print(f"R2 Score is {round(r2_score(ya,yp),2)*100}")
```
